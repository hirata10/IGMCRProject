```python
def dNdE_PL(E):
    '''
    Mrk 501
    N0: 6.6 * 10^-12 TeV^-1 cm^-2 s^-1
    alpha: 2.61
    '''
    N0 = 6.6e-12
    E0 = 1
    alpha = 2.61
    return N0 * (E/E0)**(-alpha) #* np.exp(-tau(E,z))
    
def dNdE_PL_CO(E):
    '''
    Mrk 421
    N0: 4.0 * 10^-11 TeV^-1 cm^-2 s^-1
    alpha: 2.26
    E_c: 5.1 TeV
    '''
    N0 = 4.e-11
    E0=1
    alpha = 2.26
    E_c = 5.1
    return N0 * (E/E0)**(-alpha) * np.exp(-E/E_c) # * np.exp(-tau(E,z))
```
